```python
class OFIFeatures:
    def __init__(self, df, levels=10):
        self.df = df.copy()
        self.levels = levels
        self.bid_sz_cols = [f'bid_sz_0{i}' for i in range(levels)]
        self.ask_sz_cols = [f'ask_sz_0{i}' for i in range(levels)]
        self.ofi_df = pd.DataFrame(index=self.df.index)

    def calculate_best_level_ofi(self):
        bid_change = self.df['bid_sz_00'].diff()
        ask_change = self.df['ask_sz_00'].diff()
        self.ofi_df['best_level_ofi'] = bid_change - ask_change
        return self.ofi_df['best_level_ofi']

    def calculate_multi_level_ofi(self):
        for i in range(self.levels):
            bid_col = self.bid_sz_cols[i]
            ask_col = self.ask_sz_cols[i]
            self.ofi_df[f'ofi_lvl_{i}'] = self.df[bid_col].diff() - self.df[ask_col].diff()
        return self.ofi_df[[f'ofi_lvl_{i}' for i in range(self.levels)]]

    def calculate_integrated_ofi(self):
        multi_level_ofi = self.calculate_multi_level_ofi().fillna(0)
        pca = PCA(n_components=1)
        integrated = pca.fit_transform(multi_level_ofi)
        self.ofi_df['integrated_ofi'] = integrated.flatten()
        return self.ofi_df['integrated_ofi']

    def calculate_cross_asset_ofi(self, time_col='ts_event', symbol_col='symbol'):
        if 'integrated_ofi' not in self.ofi_df.columns:
            self.calculate_integrated_ofi()

        combined = self.df[[symbol_col, time_col]].copy()
        combined['integrated_ofi'] = self.ofi_df['integrated_ofi']

        cross_asset_list = []
        grouped = combined.groupby(time_col)

        for ts, group in grouped:
            for sym in group[symbol_col].unique():
                other = group[group[symbol_col] != sym]
                cross_asset_list.append({
                    time_col: ts,
                    symbol_col: sym,
                    'cross_asset_ofi': other['integrated_ofi'].mean()
                })

        cross_df = pd.DataFrame(cross_asset_list)
        self.df = self.df.merge(cross_df, on=[symbol_col, time_col], how='left')
        return self.df['cross_asset_ofi']
```

```python
# Initialise feature constructor
ofi_feature = OFIFeatures(df)
```

```python
# Best-Level OFI
best_ofi = ofi_feature.calculate_best_level_ofi()
print("Best-Level OFI:\n", best_ofi.head(10))
```

```python
# Multi-Level OFI
multi_ofi = ofi_feature.calculate_multi_level_ofi()
print("Multi-Level OFI:\n", multi_ofi.head(10))
```

```python
# Integrated OFI
integrated_ofi = ofi_feature.calculate_integrated_ofi()
print("Integrated OFI:\n", integrated_ofi.head(10))
```

```python
# Cross Asset OFI
cross_asset_ofi = ofi_feature.calculate_cross_asset_ofi()
print("Cross-Asset OFI:\n", cross_asset_ofi.head(10))
```
